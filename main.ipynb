* Company Productivity Analysis using NumPy

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

# Load Company Productivity Data


```python
data_frame = np.loadtxt("company.txt", delimiter=",", dtype=int)

print("Company Productivity Data:\n")
print(data_frame)
```

```python
print("Shape of the array:", data_frame.shape)
print("Data type of the array:", data_frame.dtype)
```

## Indexing and Slicing

In this section, we access specific rows, columns, and individual elements from the NumPy array.

```python
print("First branch data:", data_frame[0])
print("Second column data:", data_frame[:, 1])
print("Element at row 1, column 2:", data_frame[1, 2])
```

## Total Productivity of Each Branch

Using np.sum(), we calculate the total number of products made by each branch.

```python
branch_totals = np.sum(data_frame, axis=1)
print("Total productivity of each branch:", branch_totals)
```

## Best and Worst Performing Branch

Here, we identify the branch with the highest and lowest total productivity.

```python
best_branch = np.argmax(branch_totals)
worst_branch = np.argmin(branch_totals)

print(f"Best branch: Branch {best_branch + 1} with {branch_totals[best_branch]} total products.")
print(f"Worst branch: Branch {worst_branch + 1} with {branch_totals[worst_branch]} total products.")
```

## Average Productivity

We calculate the average productivity of each branch and the overall company average using np.mean().

```python
branch_means = np.mean(data_frame, axis=1)
overall_mean = np.mean(data_frame)

for i, avg in enumerate(branch_means):
    print(f"Average productivity of Branch {i + 1}: {avg:.2f} products per employee")

print(f"\nOverall company average productivity: {overall_mean:.2f} products per employee")
```

## Salary Calculation

Assume each employee earns 7 units of money per product.  
We calculate salaries using NumPy array multiplication.

```python
salary_per_product = 7
salaries = data_frame * salary_per_product

print("Salary data:\n")
print(salaries)
```

## Column-wise and Row-wise Operations

NumPy allows us to perform operations across rows and columns using the axis parameter.

```python
print("Column-wise sum:", np.sum(data_frame, axis=0))
print("Row-wise sum:", np.sum(data_frame, axis=1))
```

## Reshaping, Flattening, and Transpose

This section demonstrates how to reshape arrays, flatten them into one dimension, and transpose rows into columns.

```python
flattened = data_frame.ravel()
transposed = data_frame.T

print("Flattened array:\n", flattened)
print("\nTransposed array:\n", transposed)
```

```python
reshaped = flattened.reshape(5, 5)
print("Reshaped array:\n", reshaped)
```

## Useful NumPy Functions

Here, we use some useful NumPy functions

```python
sample = np.array([10, 5, 20, 8, 15])

print("Sample array:", sample)
print("Index of maximum value:", sample.argmax())
print("Index of minimum value:", sample.argmin())
print("Sorted indices:", sample.argsort())
print("Values greater than 10:", np.where(sample > 10))
print("Non-zero count:", np.count_nonzero(sample))
```

## Save and Load NumPy Array

NumPy arrays can be saved into .npy files and loaded again for future use.

```python
np.save("company_data.npy", data_frame)
loaded_data = np.load("company_data.npy")

print("Loaded data from file:\n")
print(loaded_data)
```

## Compare NumPy Array with Python List

This comparison shows why NumPy is more memory efficient than Python lists for numerical data.

```python
py_list = [10, 20, 30, 40, 50]
np_arr = np.array(py_list)

python_memory = sys.getsizeof(py_list)
numpy_memory = np_arr.itemsize * np_arr.size

print("Python list memory:", python_memory, "bytes")
print("NumPy array memory:", numpy_memory, "bytes")
```

## Data Visualization

A simple line graph is plotted to visualize the productivity of employees in the first branch.

```python
first_branch = data_frame[0]
x = np.arange(len(first_branch))

plt.plot(x, first_branch, marker='o')
plt.title("Productivity of Employees in Branch 1")
plt.xlabel("Employee Index")
plt.ylabel("Products per Month")
plt.xticks(x)
plt.grid(True)
```
